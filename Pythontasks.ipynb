```python
# Employee Salary Analysis Using Python Arrays

# Import array module
from array import array

# Create Employee Salary Array
salaries = array('i', [25000, 32000, 28000, 40000, 35000])

print("Employee Salaries:")
print(salaries)

# Access Array Elements
print("\nFirst Employee Salary:", salaries[0])
print("Third Employee Salary:", salaries[2])

# Update Salary Value
salaries[2] = 30000

print("\nUpdated Salaries:")
print(salaries)

# Find Highest and Lowest Salary
highest = max(salaries)
lowest = min(salaries)

print("\nHighest Salary:", highest)
print("Lowest Salary:", lowest)

# Calculate Average Salary
average = sum(salaries) / len(salaries)

print("Average Salary:", average)

# Sort Salaries
salaries = array('i', sorted(salaries))

print("\nSorted Salaries:")
print(salaries)

# Search Employee Salary
search = 32000

if search in salaries:
    print(search, "is present")
else:
    print(search, "not found")

# Filter Salaries Above Average
average = sum(salaries) / len(salaries)

print("\nEmployees with salary above average:")

for salary in salaries:
    if salary > average:
        print(salary)
```
