```python
%%sql
SELECT
Students.student_name,
Borrow.borrow_id,
Borrow.borrow_date
FROM Students
INNER JOIN Borrow
ON Students.student_id = Borrow.student_id;
```

```python
%%sql

SELECT
Students.student_name,
Books.book_name,
Borrow_Details.quantity

FROM Students

JOIN Borrow
ON Students.student_id = Borrow.student_id

JOIN Borrow_Details
ON Borrow.borrow_id = Borrow_Details.borrow_id

JOIN Books
ON Borrow_Details.book_id = Books.book_id;
```

```python
%%sql

SELECT
Students.student_name,
Borrow.borrow_id

FROM Students

LEFT JOIN Borrow

ON Students.student_id = Borrow.student_id;
```

```python
%%sql

SELECT
Students.student_name,
COUNT(Borrow_Details.book_id) AS Total_Books_Borrowed

FROM Students

JOIN Borrow
ON Students.student_id = Borrow.student_id

JOIN Borrow_Details
ON Borrow.borrow_id = Borrow_Details.borrow_id

GROUP BY Students.student_name;
```
